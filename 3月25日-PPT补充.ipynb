# 时间序列

```python
data2 = pd.read_csv(r'C:\Users\X1\Desktop\青鸟分析temp\data_w_label.csv')
data2.head()
```

```python
data3 = data2.groupby(['stage_index', 'label'])['label'].count().unstack()
data3.head()
```

```python
for ei in data3.index:
    if np.isnan(data3.loc[ei, 0]):
        data3.loc[ei, 0]=0
    if np.isnan(data3.loc[ei, 1]):
        data3.loc[ei, 1]=0
    if np.isnan(data3.loc[ei, 2]):
        data3.loc[ei, 2]=0
    if np.isnan(data3.loc[ei, 3]):
        data3.loc[ei, 3]=0
```

```python
data3.head()
```

```python

x = data3.index
a = data3.iloc[:,0]
b = data3.iloc[:,1]
c = data3.iloc[:,2]
d = data3.iloc[:,3]
plt.figure(figsize=(30, 6))
plt.rcParams['font.sans-serif']=['SimHei'] 

plt.bar(x, a, label='0-低效阶段')
plt.bar(x, b, bottom=a, label='1-高效阶段')
plt.bar(x, c, bottom=b, label='2-偷懒阶段')
plt.bar(x, d, bottom=c, label='3-线下阶段')

plt.xticks(np.arange(0, 88, 1),rotation=90, fontsize=14)
plt.yticks(np.arange(0,160,10),fontsize=14)
plt.legend(fontsize=16)
plt.show()
```
