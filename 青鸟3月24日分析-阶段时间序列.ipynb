```python
#import MySQLdb
#conn = MySQLdb.connect(host='127.0.0.1', port=3306, user='root', passwd='123', db='edm', charset='utf8')
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
#import pandas.io.sql as sql
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv(r'C:\Users\X1\Desktop\青鸟分析temp\data_w_label.csv')
data = data.iloc[:, 1:]
data.head()
```

```python
c=['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
for i in range(14):
    data_0 = data[data['learner']==i]
    plt.figure(figsize=(35,7))
    for stage in data_0['stage_index']:
        co = stage%8;
        data_to_plt = data_0[data_0['stage_index']==stage]
        plt.scatter(data_to_plt.index, data_to_plt['label'], s=5, color=c[co])
        plt.plot(data_to_plt.index, data_to_plt['label'], color=c[co])
        plt.plot(data_to_plt.index, data_to_plt['question_correctrate'], color=c[co],alpha=0.05)
        plt.scatter(data_to_plt.index, data_to_plt['question_correctrate'], s=5, color=c[co],alpha=0.1)
    plt.axhline(y=1.0, color='black', ls='--')        
    plt.xticks([])
    plt.yticks(fontsize=16)
    plt.savefig((str)(i)+'_fig')
    plt.show()

```
